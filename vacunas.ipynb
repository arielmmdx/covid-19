```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/arielmmdx/covid-19/gh-pages/final_covid.csv')
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.columns
```

```python
df_dosis = df[['fecha','1','2']].fillna(0)
```

```python
df_dosis['fecha'] = df_dosis['fecha'].astype('datetime64')
```

```python
df_dosis.groupby(pd.Grouper(key='fecha', axis=0, freq='M', sort=True)).sum()
```

```python
df_vacunas = df[['fecha','AstraZeneca','COVISHIELD','Sinopharm','Sputnik']].fillna(0)
```

```python
df_vacunas['fecha'] = df_vacunas['fecha'].astype('datetime64')
```

```python
df_vacunas
```

```python
# applying groupby
df_vacunas = df_vacunas.groupby(pd.Grouper(key='fecha', axis=0, freq='M', sort=True)).sum()
```

```python
df_etario = pd.read_csv('https://raw.githubusercontent.com/arielmmdx/covid-19/gh-pages/final_nombre_provincias_csv.csv')
df_etario['fecha'] = df_etario['fecha'].astype('datetime64')
df_etario.loc[df_etario.provincia.str.contains('Ciudad Autónoma de Buenos Aires'),'provincia'] = 'CABA'
df_etario.loc[df_etario.provincia.str.contains('Tierra del Fuego'),'provincia'] = 'Tierra del Fuego'
```

```python
df_etario = df_etario[['provincia','18 a 39 años SIN Factores de Riesgo',
       '18 a 59 años CON Factores de Riesgo',
       '40 a 49 años SIN Factores de Riesgo',
       '50 a 59 años SIN Factores de Riesgo', '60 o más años', 'Estrategico',
       'Otros', 'Salud']].fillna(0)
```

```python
df_etario = df_etario.groupby('provincia').sum()
```

```python
from google.colab import files
df_etario.to_csv('condicion_aplicacion.csv')
files.download('condicion_aplicacion.csv')
```
